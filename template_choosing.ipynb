```python
import cv2

def create_template(image_path, save_template_path):
    # Đọc ảnh gốc
    img = cv2.imread(image_path)
    if img is None:
        print("Không tìm thấy ảnh gốc!")
        return

    print("Hướng dẫn:")
    print("1. Dùng chuột kéo một hình chữ nhật bao quanh một con thỏ.")
    print("2. Nhấn phím SPACE hoặc ENTER để xác nhận cắt.")
    print("3. Nhấn phím C để hủy và chọn lại.")

    # Mở cửa sổ cho phép người dùng chọn vùng (Region of Interest - ROI)
    roi = cv2.selectROI("Chon Template (Nhan Enter de luu)", img, fromCenter=False, showCrosshair=True)

    # roi trả về tuple: (x, y, width, height)
    x, y, w, h = int(roi[0]), int(roi[1]), int(roi[2]), int(roi[3])

    # Nếu người dùng có chọn vùng (width và height > 0)
    if w > 0 and h > 0:
        # Cắt ảnh từ ma trận numpy: img[y_start:y_end, x_start:x_end]
        template = img[y:y+h, x:x+w]
        
        # Lưu file template
        cv2.imwrite(save_template_path, template)
        print(f"Đã lưu template thành công tại: {save_template_path} với kích thước {w}x{h}")
    else:
        print("Chưa có vùng nào được chọn.")

    cv2.destroyAllWindows()

# Sử dụng hàm
image_path = "D:/AI_project/count_and_finding/Midterm-ImageSet/Midterm-ImageSet/Finding/1.jpg"
create_template(image_path, 'origin_image_finding.jpg')
```

```python
import cv2

def create_template(image_path, save_template_path):
    # Đọc ảnh gốc
    img = cv2.imread(image_path)
    if img is None:
        print("Không tìm thấy ảnh gốc!")
        return

    h, w = img.shape[:2]

    print("Hướng dẫn:")
    print("1. Dùng chuột kéo một hình chữ nhật bao quanh một con thỏ.")
    print("2. Nhấn phím SPACE hoặc ENTER để xác nhận cắt.")
    print("3. Nhấn phím C để hủy và chọn lại.")

    # 🔥 Tạo window chuẩn (không bị auto scale)
    window_name = "Chon Template (Nhan Enter de luu)"
    cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)
    cv2.resizeWindow(window_name, w, h)  # fix đúng size ảnh

    # Chọn ROI
    roi = cv2.selectROI(window_name, img, fromCenter=False, showCrosshair=True)

    # roi: (x, y, w, h)
    x, y, rw, rh = map(int, roi)

    if rw > 0 and rh > 0:
        template = img[y:y+rh, x:x+rw]

        # 🔥 Hiển thị template để kiểm tra
        cv2.namedWindow("Template Preview", cv2.WINDOW_NORMAL)
        cv2.imshow("Template Preview", template)

        # Lưu file
        cv2.imwrite(save_template_path, template)

        print(f"Đã lưu template tại: {save_template_path}")
        print(f"Kích thước template: {rw}x{rh}")
    else:
        print("Chưa có vùng nào được chọn.")

    cv2.waitKey(0)
    cv2.destroyAllWindows()


# Sử dụng
image_path = "D:/AI_project/count_and_finding/Midterm-ImageSet/Midterm-ImageSet/Finding/1.jpg"
create_template(image_path, 'no2.jpg')
```
